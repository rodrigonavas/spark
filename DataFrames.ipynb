```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, FloatType, Row
from pyspark.sql import SQLContext
```

```python
spark = SparkContext(master = "local", appName="DataFrames")
sql_context = SQLContext(spark)
```

```python
!head -n 5 {path.format('juegos.csv').replace(" ", "\ ")}
```

```python
juego_schema = StructType([
          StructField('id', IntegerType(), False),
          StructField("nombre",StringType(),False),
          StructField('anio', StringType(), False),
          StructField('temporada', StringType(), False),
          StructField('ciudad', StringType(), False)
])

juego_df = sql_context.read.schema(juego_schema).option("header", "true").csv(path.format('juegos.csv'))
juego_df.show(5)
```

```python
!head -n 5 {path.format('evento.csv').replace(" ", "\ ")}
```

```python
evento_schema = StructType([
          StructField('id', IntegerType(), False),
          StructField("evento",StringType(),False),
          StructField('deporte_id', IntegerType(), False)
])

evento_df = sql_context.read.schema(evento_schema).option("header", "true").csv(path.format('evento.csv'))
evento_df.show(5)
```

```python
!head -n 5 {path.format('deporte.csv').replace(" ", "\ ")}
```

```python
deporte_schema = StructType([
          StructField('id', IntegerType(), False),
          StructField('deporte', StringType(), False)          
])

deporte_df = sql_context.read.schema(deporte_schema).option("header", "true").csv(path.format('deporte.csv'))
deporte_df.show(5)
```

```python
!head -n 5 {path.format('paises.csv').replace(" ", "\ ")}
```

```python
paises_schema = StructType([
          StructField('id', IntegerType(), False),
          StructField('equipo', StringType(), False),
          StructField('pais', StringType(), False)          
])

paises_df = sql_context.read.schema(paises_schema).option("header", "true").csv(path.format('paises.csv'))
paises_df.show(5)
```

```python
!head -n 5 {path.format('resultados.csv').replace(" ", "\ ")}
```

```python
resultados_schema = StructType([
            StructField("id", IntegerType(), False),
            StructField("medalla", StringType(), False),
            StructField("deportista_id", IntegerType(), False),
            StructField("juego_id", IntegerType(), False),
            StructField("evento_id", IntegerType(), False)
])
resultados_df = sql_context.read.schema(resultados_schema).option("header", "true").csv(path.format('resultados.csv'))
resultados_df.show(5)
```

```python
!head -n 5 {path.format('deportista.csv').replace(" ", "\ ")}
```

```python
deportista_schema = StructType([
            StructField("id", IntegerType(), False),
            StructField("nombre", StringType(), False),
            StructField("sexo", IntegerType(), False),
            StructField("edad", IntegerType(), False),
            StructField("altura", IntegerType(), False),
            StructField("peso", IntegerType(), False),
            StructField("equipo_id", IntegerType(), False)
])
deportista_df = sql_context.read.schema(deportista_schema).option("header", "true").csv(path.format('deportista.csv'))
deportista_df.show(5)
```

```python
deportista_df = deportista_df.drop('altura')
deportista_df.printSchema()
```

```python
from pyspark.sql.functions import col

deportista_edad_df = deportista_df.select('id', 'nombre', col('edad').alias("edad_al_jugar"), 'equipo_id')
deportista_edad_df.show(5)
```

```python
deportista_edad_df.filter(deportista_edad_df.edad_al_jugar != 0).sort('edad_al_jugar').show(10)
```

```python
deportista_df.join(resultados_df, deportista_df.id == resultados_df.deportista_id, "left") \
    .join(juego_df, juego_df.id == resultados_df.juego_id, "left").join(evento_df, resultados_df.evento_id == evento_df.id) \
    .select(deportista_df.nombre.alias('Nombre deportista'), col('edad').alias('Edad al jugar'),
    'medalla', col('anio').alias('Año de juego'), col('evento').alias('Nombre de la disciplina')
    ).show(10)
```

```python
deportista_df.join(resultados_df, deportista_df.id == resultados_df.deportista_id, "inner") \
    .filter(resultados_df.medalla != 'NA') \
    .join(paises_df, deportista_df.equipo_id == paises_df.id, "inner") \
    .select(col('pais').alias('País'), col('equipo').alias('Nombre del equipo'), 'medalla') \
    .distinct().sort(col('País').asc()).show(10)
```

```python
medallistaXAnio = deportista_df \
    .join(resultados_df, deportista_df.id == resultados_df.deportista_id, "left") \
    .join(juego_df, juego_df.id == resultados_df.juego_id, "left") \
    .join(paises_df, deportista_df.equipo_id == paises_df.id, "left") \
    .join(evento_df, evento_df.id == resultados_df.evento_id, "left") \
    .join(deporte_df, evento_df.deporte_id == deporte_df.id, "left") \
    .select(
        "pais",
        "anio",
        "medalla",        
        deporte_df.deporte.alias("Nombre disciplina"),
        evento_df.evento.alias("Nombre subdisciplina"),
        deportista_df.nombre.alias("Nombre deportista")
    )
medallistaXAnio.show(10)
```

```python
medallistaXAnioCount = medallistaXAnio.filter(medallistaXAnio.medalla != 'Na') \
  .groupBy('anio', 'pais', 'Nombre subdisciplina') \
  .count().sort(col('count').desc())
medallistaXAnioCount.show()
```

```python
medallistaXAnioCount.printSchema()
```

```python
from pyspark.sql.functions import sum as _sum
from pyspark.sql.functions import avg as _avg
```

```python
medallistaXAnioCount.groupby('pais', 'anio').agg(
    _sum('count').alias('Total de medallas'),
    _avg('count').alias('Medallas promedio')
).sort('anio', col('Total de medallas').desc()).show()
```

###Registrar DataFrames como SQL

```python
resultados_df.registerTempTable('resultado')
deportista_df.registerTempTable('deportista')
paises_df.registerTempTable('paises')
```

```python
sql_context.sql('SELECT * FROM deportista').show(5)
```

```python
sql_context.sql("""
    SELECT medalla, equipo, paiS FROM resultado r JOIN deportista d 
    ON r.deportista_id = d.id
    JOIN paises p ON p.id = d.equipo_id 
    WHERE medalla <> "NA"
    ORDER BY pais DESC
""").show(10)
```
